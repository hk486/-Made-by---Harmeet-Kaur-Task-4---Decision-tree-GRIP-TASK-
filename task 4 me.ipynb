# Splitting Dataset into Training and Testing

```python
x =data.drop(['Species'],axis=1)
```

```python
y = data['Species']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=0)
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini' , random_state=0)
clf_gini.fit(x_train, y_train)
```

```python
prediction_y=clf_gini.predict(x_test)
```

```python
print(prediction_y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("The Accuracy score is:",accuracy_score(y_test,prediction_y)*100)
```

```python
from sklearn import tree
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm']
```

```python
plt.figure(figsize=(24,20))
tree.plot_tree(clf_gini.fit(x,y),feature_names=features,filled=True,precision=3,proportion=True,rounded=True)
plt.show()
```
